```python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error, median_absolute_error, r2_score 
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC, NuSVC, LinearSVC, OneClassSVM, SVR, NuSVR, LinearSVR
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
import seaborn as sns
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
%matplotlib inline 
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
df_int = le.fit_transform(df['City'])
df['City'] = df_int
df_int = le.fit_transform(df['Gender'])
df['Gender'] = df_int
df_int = le.fit_transform(df['Illness'])
df['Illness'] = df_int
```

```python
df.dtypes
```

```python
target = df['target']
data_X_train, data_X_test, data_y_train, data_y_test = train_test_split(
    df, target, test_size=0.2, random_state=1)
```

```python
df = df.head(1000)
```

```python
svr_1 = LinearSVC(dual=False)
svr_1.fit(data_X_train, data_y_train)
```

```python
data_y_pred_1 = svr_1.predict(data_X_test)
```

```python
accuracy_score(data_y_test, data_y_pred_1)
```

```python
f1_score(data_y_test, data_y_pred_1, average='micro')
```

```python
f1_score(data_y_test, data_y_pred_1, average='macro')
```

```python
f1_score(data_y_test, data_y_pred_1, average='weighted')
```
