```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Admission_Predict_Ver1.1.csv')
```

Посмотрим колчиество пропущеных значений

```python
df.isna().sum()
```

т.к. нету пустых значений, добавим их

```python
df = df.mask(np.random.random(df.shape) < 0.01)
```

```python
df.isna().sum()
```

Удалим искусственно созданные нулевые значения

```python
df = df.dropna()
```

```python
df.isna().sum()
```

Создадим парные диаграммы

```python
sns.pairplot(df)
```

Отобразим корреляционную матрицу

```python
plt.figure(figsize=(8, 5))
plt.title('Матрица корреляций')
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

Из полученной матрицы видно, что больше всего на успех поступления влияют: CGPA(Средний балл по баллам), GRE Score(Оценка по тесту Graduate Record Examinations) и TOEFL Score(Оценка по тесту TOEFL).
